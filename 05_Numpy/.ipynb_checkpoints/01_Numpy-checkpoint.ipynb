```python

np.random.seed(10)
mat = np.random.randint(1,5,(2,3))
print(mat)
print(mat.argmax()) # it will flatten it as a row major 1D vec and give the index of max. 
```

```python
mat[1]
```

```python
import numpy as np

np.random.seed(10)
mat = np.random.randint(1,5,(2,3))
print("Matrix:\n", mat)

# Get the index of the max element in 1D flattened form
max_index = mat.argmax()

# Convert to 2D (row, col) index
row, col = np.unravel_index(max_index, mat.shape)
print("Max element is at row:", row, "column:", col)
```
