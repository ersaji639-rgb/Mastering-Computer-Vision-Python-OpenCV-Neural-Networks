```python
import numpy as np

parent = np.array([1, 2, 3])
child = parent.copy()   # deep copy, independent of parent

child[0] = 100

print("Parent:", parent)  # Output: [1 2 3]
print("Child: ", child)   # Output: [100 2 3]
```

```python
import numpy as np

arr = np.arange(10)

# First slice, then make a copy
arr_slice = arr[1:4].copy()  

# Modify the slice
arr_slice = arr_slice + 100

print("Original array:", arr)      # unchanged
print("Modified slice:", arr_slice)
```
