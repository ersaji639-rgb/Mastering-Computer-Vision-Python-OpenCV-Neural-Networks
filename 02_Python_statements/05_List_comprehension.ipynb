```python
num_lst = [] 
for n in range(11):
    if n%2 ==0:
        num_lst.append(n**2)

print(num_lst)
    
```

### List Comprehension:

```python
[w.upper() for w in st.split()]
```

```python
[n**2 for n in range(11)]
```

```python
[n**2 for n in range(11) if n%2 == 0]
```

```python
[n**2 if n%2 == 0 else 'ODD' for n in range(11) ]
```

```python
[n**2 if n%2 == 0 else n for n in range(11) ]
```

```python
[n**2 if n%2 == 0 else f'{n} is ODD' for n in range(11) ]
```

```python
[y for y in [n**2 for n in range(11) if n%2 == 0 ] if y > 30] # nested list comprehension
```

```python
[y for y in [n**2 for n in range(11) if n%2 == 0 ] ]
```

```python
[w.upper() for w in "Bengalore"]
```

```python
[f"{w.upper()} has {len(w)} letters " for w in "Bengalore Weather ME".split()]
```
