```python
# enumerate 
idx_count = 0

for let in "abcd":
    print(f"{let}")
```

```python
# enumerate 
idx_count = 0

for let in "abcd":
    print(f"{let} at index {idx_count}")
    idx_count += 1
```

```python
list(enumerate('abcd'))
```

```python
# enumerate  in action


for idx_count, let in enumerate("abcd"):
    print(f"{let} at index {idx_count}")
   
```

```python
# Zip 
l1 = [11,2,12,45,65,6]
l2 = [1,'a',4,'saji',7,44,98,9]
zip(l1,l2)
```

```python
list(zip(l1,l2)) # gives tuple 
```

```python
for i1,i2 in zip(l1,l2):
    print(f"For this tuple item 1: {i1} and item 2: {i2}")
```
