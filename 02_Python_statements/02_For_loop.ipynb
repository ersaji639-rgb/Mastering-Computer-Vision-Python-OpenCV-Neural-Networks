###### Q2 
Given a string  create a dictionary with two keys "vowels" and "consonants", Ignore non alphabatic chars.

```python
#st = input('Enter your string: ')
st = "hai ..hsi..7 you"
vowels = []
consonants = []
for ch in st.lower():
    if ch.isalpha():
        if ch in 'aeiou':
            vowels.append(ch)
        else:
            consonants.append(ch)

print(f"Vowels : {vowels} \n Consonants :{consonants}")
    
```

```python
vowels = ['a', 'i', 'i', 'o', 'u']
consonants = ['h', 'h', 's', 'y']

result = {
    "Vowels": vowels,
    "Consonants": consonants
}

print(result)
```

```python
for num in range(1, 11):
    if num == 2:
        print("The num 2 is Even prime")
    elif num % 2 == 0:
        print(f"The num {num} is even number")
    else:
        isPrime = True
        for i in range(2, num):
            if num % i == 0:
                isPrime = False
                break
        if isPrime and num != 1:
            print(f"The number {num} is odd prime")
        else:
            print(f"The number {num} is odd")
```

```python
for num in range(1, 11):
    if num == 2:
        print("The num 2 is Even prime")
    elif num % 2 == 0:
        print(f"The num {num} is even number")
    else:
        isPrime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                isPrime = False
                break
        if isPrime and num != 1:
            print(f"The number {num} is odd prime")
        else:
            print(f"The number {num} is odd")
```
