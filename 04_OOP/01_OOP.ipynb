### class 

*A class is like a blueprint for creating objects.
It defines what attributes (variables) and methods (functions) the objects will have.

```python
class Dog:
    def __init__(self, name, age):   # constructor
        self.name = name
        self.age = age

    def bark(self):   # method
        print(f"{self.name} is barking!")
```

### Object/Instances: 
* An object (or instance) is a real thing created from the class.
Each object has its own data.

```python
dog1 = Dog("Tommy", 3)   # object (instance)
dog2 = Dog("Lucy", 5)    # another object
```

### Attributes
* Attributes are variables that belong to the object.

```python
print(dog1.name)  # Tommy
print(dog2.age)   # 5
```

### Methods: 
Methods are functions inside a class that describe what the object can do.

```python
dog1.bark()  # Tommy is barking!
dog2.bark()  # Lucy is barking!
```

* Putting it all together:

* Class = blueprint (like Dog).

* Object / Instance = actual thing created (like dog1, dog2).

* Attributes = object variables (like name, age).

Methods = object functions (like bark).

** Eg: 
Class = Car blueprint.

Object = Actual car built (Car1, Car2).

Attributes = color, speed, brand of that car.

Methods = drive(), brake(), honk().

```python
class Car:
    def __init__(self, name, manf_year):
        self.name = name
        self.manf_year = manf_year

    def car_details(self):
        print(f"your car name is {self.name} and manufacturing year is {self.manf_year}")
        
```

```python
car1 = Car('TATA HARRIER MATT BLACK', 2025) # creating instances. 
car1.car_details()
```

###  Vechile class :

```python
class Vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def info(self):
        return f"Brand: {self.brand}, Speed: {self.speed} km/h"
```

#### Sub class for each vechile :

```python
class Car(Vehicle):
    def __init__(self, brand, speed, doors):
        super().__init__(brand, speed)  # call parent constructor
        self.doors = doors              # Car-specific attribute

    def car_info(self):
        return f"{self.info()}, Doors: {self.doors}"
```

```python
class Bus(Vehicle):
    def __init__(self, brand, speed, capacity):
        super().__init__(brand, speed)
        self.capacity = capacity

    def bus_info(self):
        return f"{self.info()}, Capacity: {self.capacity} passengers"
```

```python
class Train(Vehicle):
    def __init__(self, brand, speed, wagons):
        super().__init__(brand, speed)
        self.wagons = wagons

    def train_info(self):
        return f"{self.info()}, Wagons: {self.wagons}"
```

```python
class Flight(Vehicle):
    def __init__(self, brand, speed, altitude):
        super().__init__(brand, speed)
        self.altitude = altitude

    def flight_info(self):
        return f"{self.info()}, Altitude: {self.altitude} m"
```

```python
car1 = Car("Toyota", 180, 4)
bus1 = Bus("Volvo", 100, 50)
train1 = Train("Bullet", 300, 10)
flight1 = Flight("Boeing", 900, 10000)

print(car1.car_info())
print(bus1.bus_info())
print(train1.train_info())
print(flight1.flight_info())
```

```python
veh1 = Vehicle("BMW",120)
print(veh1.info())
```

```python
class Circle:
    pi = 22/7
    
    def __init__(self, radius=1):  # default radius = 1
        self.radius = radius
        self.circumference = 2 * Circle.pi * radius  # here we can use self.pi as well. 
        self.area = self.pi*self.radius*self.radius

    # Define method at the same indentation as __init__
    def setRadius(self, new_radius):
        self.radius = new_radius
        self.circumference = 2 * Circle.pi * new_radius
        self.area = self.pi*self.radius*self.radius

    # Method to get the circumference. 

    def getCircumference(self):
        return self.radius*self.pi*2

    def getArea(self):
        return self.pi*self.radius*self.radius
```

```python
c1 = Circle() 
```

```python
c1.setRadius(new_radius= 6)
```

```python
c1.getCircumference()
```

```python
c1.getArea()
```

```python
c1.area
```

### Inheritance: 
* We can create new classes using the classes which are  already defined.

* Derived class: THis is the newly formed class.

* Base class: Classes that we derive from. 

```python
class Animal:
    def __init__(self):
        print("Animal Created!!")

    def whoAmI(self):
        print("Animal")
    def eat(self):
        print("Animal is eating")
```

```python
a1 = Animal()
```

```python
a1.eat()
```

```python
a1.whoAmI()
```

```python
class Dog(Animal):  # this class is going to inherit all the prop of class Animal. 
    def __init__(self):
        Animal.__init__(self)
        print("Animal is Dog")

    def whoAmI(self):
        print("Iam a dog") # I have over writtenthe old inherited method 

    def bark(self):
        print("wof")
```

```python
d1 = Dog()
```

```python
d1.eat()
```

```python
d1.whoAmI()
```

```python
a1.whoAmI()
```

```python
d1.bark()
```

#### Special methods:

```python
class Novel:
    def __init__(self, title, author, pages, color):
        print("A novel has been created!!")
        self.title = title
        self.author = author
        self.pages = pages
        self.color = color

    def __str__(self):
        return f"This book '{self.title}' is written by {self.author} has {self.pages} pages with color {self.color}" 

    def __len__(self):
        return self.pages

    def __del__(self):
        print("Novel has been deleted")
```

```python
class BankAccount:
    def __init__(self,balance = 100):
        self.balance = balance

    def deposit(self, amount):
        self.balance = self.balance +amount
        return f"the updated balance is {self.balance}"
    def widthdraw(self,amount):
        if amount < self.balance:
            self.balance -= amount
            return f"The updated balance is {self.balance}"
        else:
            return "Insuf fund"
        
        

class SavingsAccount(BankAccount):
    def __init__(self,balance = 100, interest_rate = 0.04):
        super().__init__(balance) # super allows a method from the parent class inside a child class. 
        self.interest_rate = interest_rate

    def apply_interest_rate(self):
        self.balance = self.balance + self.balance*self.interest_rate
        return self.balance
```

```python
sb1 = SavingsAccount(balance=10000,interest_rate=0.06)
```

```python
sb1.balance
```

```python
sb1.deposit(amount=100)
```

```python
sb1.balance
```

```python
sb1.widthdraw(5000)
```

```python
sb1.widthdraw(10000)
```

```python
sb1.balance
```

```python
sb1.apply_interest_rate()
```
