### f-strings (formatted strings literal) 

```python
name ='saji'
print(f'Name of the person is {name.capitalize()}')
```

```python
# captalize only the first char. 
"sAJI".capitalize()   # → "Saji"
"saji".capitalize()   # → "Saji"
"SAJI".capitalize()   # → "Saji"
```

```python
age = 28
print(f"Hai Saji your age is {age!r}")
```

```python
a = "\nH\na\ni"
print(f"You entered {a!r}")
```

#### Padding and precision of floating point numbers:
%5.2 : 5 --> minimum number of character the string should have 
%0.2f --> how many number of decimal

```python
print('Floating point number : %5.2f'%(13.644))
```

```python
print('Floating point number : %5.2f'%(13979765.644))
```

```python
print('Floating point number : %25.2f'%(13.644))
```

```python
print('Floating point number : %25.34f'%(13.644))
```
