A person's health insurance premium was 2878902 INR inclusive of GST (%18)
Calculate new premium value after slashing of GST 0%.

```python
oldpremiumAmount = int(input("Enter the old premium amount: "))  # since by default it is a string 
oldtaxPercent = 18 # in percentage 
base_amt = oldpremiumAmount*(1-oldtaxPercent/100)
newtaxPercent = 0
newPremium  = base_amt - (newtaxPercent/100) * base_amt
print("The new premium amount is {} INR".format(newPremium)) 

```
