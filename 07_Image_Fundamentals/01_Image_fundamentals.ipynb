### Using matplot lib library

```python
cim = plt.imread('tulips.jpg') 
cim = cim.copy()
plt.imshow(cim)
```

```python
R = cim[:,:,0]
G = cim[:,:,1]
B = cim[:,:,2]
```

```python
plt.figure(1)

plt.subplot(2,3,1)
plt.imshow(cim)
plt.xticks([])
plt.yticks([])

plt.subplot(2,3,2)
plt.imshow(cim)
plt.xticks([])
plt.yticks([])


plt.subplot(2,3,3)
plt.imshow(cim)
plt.xticks([])
plt.yticks([])

plt.subplot(2,3,4)
plt.imshow(R,cmap = 'gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2,3,5)
plt.imshow(G,cmap = 'gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2,3,6)
plt.imshow(B,cmap = 'gray')
plt.xticks([])
plt.yticks([])


plt.show()
```

```python
R = cim[:,:,0]
G = cim[:,:,1]
B = cim[:,:,2]
```

```python
R[150:450,150:450] = 100
G[150:450,150:450] = 50
B[150:450,150:450] = 80
```

```python
plt.figure(1)

plt.subplot(2,3,1)
plt.imshow(cim)
plt.xticks([])
plt.yticks([])

plt.subplot(2,3,2)
plt.imshow(cim)
plt.xticks([])
plt.yticks([])


plt.subplot(2,3,3)
plt.imshow(cim)
plt.xticks([])
plt.yticks([])

plt.subplot(2,3,4)
plt.imshow(R,cmap = 'gray')
plt.xticks([])
plt.yticks([])
plt.title('red channel')


plt.subplot(2,3,5)
plt.imshow(G,cmap = 'gray')
plt.xticks([])
plt.yticks([]) 
plt.title('red channel')

plt.subplot(2,3,6)
plt.imshow(B,cmap = 'gray')
plt.xticks([])
plt.yticks([]) 
plt.title('blue channel')


plt.show()
```
