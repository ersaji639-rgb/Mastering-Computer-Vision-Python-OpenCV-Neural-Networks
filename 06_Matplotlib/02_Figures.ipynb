```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
print(a)
b = a**4
print(b)
```

```python
x = np.arange(10)
y = 2*x
```

#### Create a fig using OOP method. 

```python
# creating a blank canvas:
fig = plt.figure()
fig
```

```python
fig = plt.figure(figsize=(10,10))
```

```python
fig = plt.figure(figsize=(10,10))

# Add set of axes to the figure 
axes = fig.add_axes([0,0,1,1])
```

```python
# creating a canvas
fig = plt.figure(figsize=(4,4))

# Add set of axes to the figure 
axes = fig.add_axes([0,0,1,1])

# plot on the blank space 
axes.plot(x,y)
plt.show()
```

```python
# creating a canvas
fig = plt.figure(figsize=(4,4))

# Add set of axes to the figure 
axes = fig.add_axes([0,0,1,1])

# plot on the blank space 
axes.plot(a,b)
plt.show()
```

```python
fig = plt.figure(figsize=(5,3), dpi=200)
# we need to use set_
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlabel('x')
axes1.set_ylabel('y')


axes2 = fig.add_axes([0.2,0.5,0.5,0.4])
axes2.plot(a,b)
axes2.set_xlabel('x_inset')
axes2.set_ylabel('y_inset')
axes2.set_xlim(4.5,6.6)

fig.savefig('samplefig2.png',bbox_inches = 'tight')
plt.show()
```
