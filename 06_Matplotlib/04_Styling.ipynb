```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.arange(10)
y = x**2
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label = "x squared") 
ax.plot(x,x**3, label = "x cubed")
#ax.legend(loc = 'upper left')
ax.legend(loc = (1.1,0.5))
plt.show()
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label = 'x squared',linestyle = '--',color = '#B37171',marker  = 's',mec = 'black',mew = 2, mfc = 'green',ms = 12) 
ax.plot(x,x**3, label = "x cubed",ls = '-.',lw = 6,color = 'red',marker = '8')
#ax.legend(loc = 'upper left')
ax.legend(loc = (1.1,0.5))
plt.show()
```
