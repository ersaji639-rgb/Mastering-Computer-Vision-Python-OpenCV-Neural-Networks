```python
fig, axes1 = plt.subplots(nrows=1, ncols=2)
axes1[0].plot(a,b)
axes1[-1].plot(b,a)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0, 1].plot(a, b)
axes[0, 1].set_xlabel('x2')
axes[0, 1].set_ylabel('y2')
axes[0, 1].set_title('plot1')

axes[1, 0].plot(a, b**0.2)

fig.suptitle('Figure level title', fontsize=10)
fig.supxlabel('X')
fig.supylabel('Y')


#fig.subplots_adjust(left=5, right=2)
# use wspace, hspace 
fig.subplots_adjust(wspace=0.9, hspace=0.7)

plt.tight_layout(rect=[0, 0, 1, 0.95])  # leave space for suptitle

plt.savefig('Sample3.png')  # save before showing
plt.show()
```
