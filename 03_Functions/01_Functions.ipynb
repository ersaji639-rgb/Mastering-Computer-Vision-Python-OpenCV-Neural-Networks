```python
def greet_fun(name):
    # This prints the message directly to the console (side effect).
    # It does NOT give back a value to the caller.
    print("Hello " + name)  
    
    # This returns the message as a value that the caller can capture
    # in a variable or use in further computation.
    return ("Hello " + name)
```

```python
# Call the function
result = greet_fun("Saji")

# 'print' inside the function shows:
# Hello Saji

# 'return' gives back "Hello Saji", which we stored in result
print("Returned value:", result)
```

```python
def check_even_list(num_list):
    for n in num_list:
        if n%2 == 0:
            return True # once it is correct it exits the function.
        else:
            pass # if odd dont do any thing and go to next number.

    return False # if never encountered any even 
```

```python
def check_even_list(num_list):
    # Handle the case where the list is empty
    if not num_list:   # same as len(num_list) == 0
        return "List is empty"

    for n in num_list:
        if n % 2 == 0:
            return True  # exits immediately when an even is found
        else:
            pass         # skip odd numbers
    
    # If no even number is found
    return False
```

```python
print(check_even_list([]))        # List is empty
print(check_even_list([1,3,5]))   # False
print(check_even_list([1,4,5]))   # True
```

```python
def  op_oddnumbers(num_lst):
    odd_num = []
    for num in num_lst:
        if num % 2 != 0:
            odd_num.append(num)
    return odd_num
```

```python
n = [1,2,3,5]
op_oddnumbers(n)
```

* [expression for item in iterable if condition]


```python
# Using list comprehension:  print all odd numbe in lst 
num_lst = [1, 2, 3, 4, 5, 6, 7]
odd_numbers = [x for x in num_lst if x % 2 != 0]
print(odd_numbers)
```

```python
def op_oddnumbers_alt(num_lst):
    return [x for x in num_lst if x % 2 != 0]
```

```python
op_oddnumbers_alt(num_lst)
```

* [expression_if_true if condition else expression_if_false for item in iterable]


```python
num_lst = [1, 2, 3, 4, 5, 6, 7]
result = [x if x % 2 != 0 else "even" for x in num_lst]
print(result)
```

#### Tuple unpacking:

```python
stock_prices = [('AAPL',200),('MSFT',400),('TSL',250),('RECL',200)]
```

```python
for item in stock_prices:
    print(item)
```

```python
for s, p in stock_prices:
    print(f"Stock {s}, price {p}")
```

```python
# Stock of the month based on the cost: 
def stock_check(st_lst):
    current_max = 0
    stock_of_month = ''
    for s, p in st_lst:
        if p > current_max:
            current_max = p
            stock_of_month = s
    return f"Stock of the month is {stock_of_month} at Rs {current_max}"
```

```python
stock_check(st_lst= stock_prices)
```
