### *args 
* When a function starts with * it allows an arbitary function takes them as a tuple
* We can use, when we cont know how many input arguments will be there. 

```python
def my_fun(*arg):
    return arg
```

```python
my_fun(1,2,3)
```

```python
def my_fun(*arg):
    return sum(arg)*0.5
```

```python
my_fun(1,2,3)
```

```python
def my_fun(*arg):
    return arg
```

```python
my_fun('Appe','tom')
```

```python
def my_fun(*arg):
    return  ' and ' .join(arg)
```

```python
my_fun('Appe','tom')
```

### **kwargs

```python
def myFun(**kwargs):
    if 'fruit' in kwargs:
        print(f"My fav fruit is {kwargs['fruit']}")
    elif 'veg' in kwargs:
        print(f"My fav Veg is {kwargs['veg']}")
    else:
        print("I dont like this fruit")
```

```python
myFun(fruit = 'apple')
```

```python
myFun(veg = 'Ladys finger')
```

```python
def myFun(*args,**kwargs):
    if 'fruit' and 'juice' in kwargs:
        print(f"i like {' and '.join(args)}")
```

```python
myFun('eggs','Grape',fruit = 'mangos', juice = 'Orange')
```
