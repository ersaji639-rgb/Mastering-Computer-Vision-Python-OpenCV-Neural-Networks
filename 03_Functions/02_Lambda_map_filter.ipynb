### Lambda Expressions: Oneliner anonymous function (One time use function without any name)

### Map:
* Allows you to map a function to some iterable object (Eg: list, str etc )

```python
def square(num):
    return num**2
```

```python
square(2)
```

```python
lst = [1,2,3,4,5,6]
# To square: 
op = []
for num in lst:
    op.append(num**2)
print(op)
```

```python
# using list comprehension:
[square(n)  for  n in lst] 
```

```python
map(square, lst) # Generator 
```

```python
list(map(square, lst)) # we need to cast as a list. 
```

```python
def manipulate_name(nm):
    if len(nm) % 2 == 0:
        return 'even'
    else:
        return nm.upper()
```

```python
lst = ['SAJI', 'John', 'Lebron','Giannis','Kevin']
```

```python
list(map(manipulate_name,lst))
```

### filter:
* this returns an iterator yielding those items for which the function returns true.
* Function should return boolean value.

```python
def even_length(nm):
    return len(nm)%2 == 1 
```

```python
even_length('SajiM')
```

```python
list(filter(even_length,lst))
```

```python
s = ['Saji','Mechanical','Aero','Boxing']
# grab the first char of all str 
# reverese all word in list 
# make a list whose elements are cube of even numbers. 
# take a list of numbers and return list of palindrome [121,343, 787,332] == > [121,343,787]
```

```python
s
```

```python
# grab the first char of all str 
first_ch = list(map((lambda ch : ch[0]), s))
```

```python
print(first_ch)
```

```python
# reverese all word in list 
list(map((lambda word : word[::-1]),s))
```

```python
# make a list whose elements are cube of even numbers. using lambda 
lst1 = [1, 2, 3, 4, 5]

result = list(map(lambda x: x**3, filter(lambda x: x % 2 == 0, lst1)))
print(result)

```

```python
 list(filter(lambda x: x % 2 == 0, lst1))
```

```python
# take a list of numbers and return list of palindrome [121,343, 787,332] == > [121,343,787] 
lst2 = [121,343, 787,332]

list(filter(lambda word : str(word) == str(word)[::-1], lst2))

```

```python
# Using list comprehension: 
lst2 = [121, 343, 787, 332]

palindromes = [x for x in lst2 if str(x) == str(x)[::-1]]
print(palindromes)
```

```python
lst2 = [121, 343, 787, 332, -121, -787]

palindromes = [x for x in lst2 if str(abs(x)) == str(abs(x))[::-1]]
print(palindromes)
```
